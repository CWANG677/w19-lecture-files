# Matplotlib Recipes

The default look you get from the matplotlib tutorial https://matplotlib.org/users/pyplot_tutorial.htm

```python
import matplotlib.pyplot as plt
%matplotlib inline

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.savefig("basic.png", format='png', bbox_inches='tight')
```

With titles and labels

```python
# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.title('$y=Ax$',size=15)
plt.savefig("title.png", format='png', bbox_inches='tight')
```

With seaborn

```python
from IPython.display import set_matplotlib_formats
import seaborn as sns
%matplotlib inline

sns.set_style("darkgrid")
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.title('$y=Ax$',size=15)
plt.savefig("sea.png", format='png', bbox_inches='tight')
```
